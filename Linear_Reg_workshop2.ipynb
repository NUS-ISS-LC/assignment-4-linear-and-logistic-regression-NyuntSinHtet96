# Linear Regression Example 

```python
'''In this example, we will build a linear regression with two independent variables.  This linear regression model is also called multiple linear regression.
1) Load the dataset 'OmniPower.csv'
2) Create a Linear Regression Model with two independent variables (Price and Promotion), one dependent variable (Sale)
3) Provide input values and predict the Sale
4) What is the R square value?'''
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('OmniPower.csv')
df.head(10)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
features = df[['Price','Promotion']]
Sales = df['Sales']
print(features)
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,Sales,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(x_train,y_train)
print(linReg.intercept_)
print(linReg.coef_)
```

```python
y_pred = linReg.predict(x_test)
print(y_pred)
```

```python
print(linReg.predict([[60,200]]))
```

```python
from sklearn.metrics import r2_score
print('r2_score: ',r2_score(y_test,y_pred))
print('score: ',linReg.score(x_test,y_test))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(x_test['Price'],y_test,color='green',label='Actual sales')
plt.scatter(x_test['Promotion'],y_test,color='pink',label='Actual sales')
plt.xlabel("Price")
plt.ylabel("Sale")
plt.title("Price vs Sale")
plt.legend()
plt.grid(True)
plt.show()
```
